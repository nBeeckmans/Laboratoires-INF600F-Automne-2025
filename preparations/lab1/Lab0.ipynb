## Exercice 3
Complétez la fonction `compare` pour comparer 2 nombres.

```python
def compare(nb1, nb2):
    """
    Si nb1 < nb2, retourne -1
    Si nb1 > nb2, retourne 1
    Si nb1 == nb2, retourne 0
    """
    
    # Si nb1 < nb2, retourne (-1, 'inferieur')
    if nb1 < nb2:
         return (-1, 'inferieur') 
    
    # Si nb1 > nb2, retourne (1, 'superieur')
    if nb1 > nb2:
         return (1, 'inferieur') 
    
    # Si nb1 == nb2, retourne (0, 'egal')
    
    return (0, 'egal')

a = 10; b = 25
print('{:<4}'.format(str(a)), '{:<4} :'.format(str(b)), compare(a, b))
a = 111; b = 92
print('{:<4}'.format(str(a)), '{:<4} :'.format(str(b)), compare(a, b))
a = 1000; b = 1000
print('{:<4}'.format(str(a)), '{:<4} :'.format(str(b)), compare(a, b))
```

## Exercice 7
Appelez la fonction `mystere`, storez les valeurs retournez dans 3 différentes variables et affichez les.

> Vous pouvez voir que la fonction utilise des paramètres par défaut,
> essayez de donner à la fonction les paramètres `a` et `c` seulement.
> Comme ça, on pourra utiliser la valeur par défaut de `b`.

```python
from random import random

def mystere(a, b = 2, c = 3):
    if b != 0:
        a = a ** 2
        a = a // b
        c /= b
    else:
        a = round(random())
        c = random() * 10
        
    return a, b, c

print(mystere(a=3, b=0))
```

```python
# Python fait usage de `deep` et `shallow copy` dépendemment du type de la variable passé.

print('Passé en paramètre un integer:')
a = 5

def add(nb, add):
    nb += add
    return nb

print(a)
print(add(a, 5))
print(a)

print('---')

print('Passé en paramètre une liste (objet mutable):')
list_a = [5, 10]

def list_add(list_nb, add):
    for i in range(0, len(list_nb)):
        list_nb[i] += add
    return list_nb

print(list_a)
print(list_add(list_a, 5))
print(list_a)

print('---')
print('Erreur typique que je fais souvent:')
test = [[0, 0, 0]] * 3
print(test)
test[0][0] = 1
print(test)

print('---')
print('Solution pour cette erreur:')
test = []

for i in range(0,3):
    test.append([0, 0, 0])

print(test)
test[0][0] = 1
print(test)
```

```python
from skimage import io
from skimage import color
import numpy as np
import matplotlib.pyplot as plt


image = io.imread("sunflower.jpg")
imageAsArray = np.array(image)
print(imageAsArray.shape)

def create_square(image, i_min, i_max, j_min, j_max): 
    min_value = image.min()
    max_value = image.max()
    for i in range(i_min, i_max):
        for j in range(j_min, j_max):
            image[i][j][0] = max_value # rouge
            image[i][j][1] = min_value # vert
            image[i][j][2] = max_value # bleu
    return image
    

# utiliser un jpeg pour tester !
def image2gray(image):
    imageRGB = image;
    if image.shape[2] == 4:
        imageRGB = color.rgba2rgb(image)

    return color.rgb2gray(imageRGB)
    

# peut-etre ici ?
plt.subplot(1,2,2)
plt.imshow(image2gray(imageAsArray), cmap="gray")

plt.subplot(1,2,1)
plt.imshow(create_square(imageAsArray, 100, 120, 190, 210))

plt.show()
```
