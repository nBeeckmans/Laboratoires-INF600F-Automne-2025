```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plotSin(amplitude=1, pulsation=1, phase=0):
    x = np.linspace(0, 2*np.pi, 512)
    def f(x):
        return amplitude * np.sin(x * pulsation + phase)
    plt.plot(x,f(x))

plotSin()
plotSin(amplitude=2)
plotSin(pulsation=2)
plotSin(phase=np.pi)
```
